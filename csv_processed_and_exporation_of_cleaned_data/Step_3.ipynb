```python
# Regression Plots
```

```python
edu_cols=['HSGradPer', 'BADegPer','AdvDegPer']
```

```python
edu_cols
```

```python
edu_data=df.loc[:,['LifeExp2018','Region'] +  edu_cols]
```

```python
edu_data
```

```python
edu_data_num=df.loc[:,['LifeExp2018']+edu_cols]
```

```python
edu_data_num
```

```python
# compute the correlation matrix
corr_matrix_edu=edu_data_num.corr()
```

```python
corr_matrix_edu
```

```python
# heatmap showing the correlation of the edu variables and life expectancy
```

```python
fig, ax0=plt.subplots()
sns.heatmap(corr_matrix_edu, annot=True, cmap='YlGnBu', ax=ax0)
#margin bigs
#b, t= plt.ylim ()
# plt.ylim(b+0.5, t-.5)
plt.yticks(rotation=45)
ax0.set_title('correlation of Education variables and Life Expectancy')
plt.savefig('heatmap_edu.png')
plt.show()
```

```python
# Regression plots for the education variables
```

```python
Corr_HSGradper=corr_matrix_edu.loc['LifeExp2018', 'HSGradPer']
Corr_BADegPer=corr_matrix_edu.loc['LifeExp2018', 'BADegPer']
Corr_AdvDegPer=corr_matrix_edu.loc['LifeExp2018', 'AdvDegPer']
```

```python
Corr_HSGradper
```

```python
Corr_BADegPer
```

```python
Corr_AdvDegPer
```

```python
fig, (ax1, ax2, ax3)=plt.subplots(nrows=1,ncols=3, sharey=True, figsize=(17,4))
#High school Degree
s1=sns.regplot(x='HSGradPer', y='LifeExp2018', color='Purple', data=df, ax=ax1)
s1.set(title='Corr: {:.4f}'.format(Corr_HSGradper), xlabel='Highschool Graduates (%)', ylabel='Life Expectancy')

#Bachelor Degree
s2=sns.regplot(x='BADegPer', y='LifeExp2018', color='Purple', data=df, ax=ax2)
s2.set(title='Corr: {:.4f}'.format(Corr_BADegPer), xlabel='Bachelor Degree Graduates (%)', ylabel='')

# Advanced degree
s3=sns.regplot(x='AdvDegPer', y='LifeExp2018', color='Purple', data=df, ax=ax3)
s3.set(title='Corr: {:.4f}'.format(Corr_AdvDegPer), xlabel='Advanced Degree Graduates (%)', ylabel='')

fig.suptitle ('Life Expectancy per state (2018)')
plt.savefig('regplots_edu.png')
plt.show()
```

```python
# Crime variable
murder_cols=['MurderNMTot', 'MurderTot','GunMurderTot','GunOwnPer','MurderNMRate','MurderRate','GunMurderRate']
```

```python
murder_data=df.loc[:,['LifeExp2018', 'Region']+ murder_cols]
```

```python
murder_data
```

```python
murder_data_num=df.loc[:,['LifeExp2018']+murder_cols]
```

```python
murder_data_num
```

```python
corr_matrix_murder=murder_data_num.corr()
```

```python
corr_matrix_murder
```

```python
fig, ax0=plt.subplots()
sns.heatmap(corr_matrix_murder, annot=True, cmap='YlGnBu', ax=ax0)
# margin bugs
#b, t=plt.ylim()
#plt.ylim(b+0.5, t-.5)
plt.yticks(rotation=45)
ax0.set_title('Correlation of Murder Variables and Life Expectancy')
plt.savefig('heatmap_murder.png')
plt.show()
```

```python
corr_MurderNMTot=corr_matrix_murder.loc['LifeExp2018','MurderNMTot']
corr_GunMurder=corr_matrix_murder.loc['LifeExp2018','GunMurderTot']
corr_MurderTot=corr_matrix_murder.loc['LifeExp2018','MurderTot']
```

```python
fig, (ax1, ax2, ax3)=plt.subplots(nrows=1,ncols=3,sharey=True,figsize=(17,4))
# Murder and non-negligent Manslaughte
s1=sns.regplot(x='MurderNMTot', y='LifeExp2018', data=df, ax=ax1)
s1.set(title='Corr: {:.4f}'.format(corr_MurderNMTot), xlabel='Murder and non-negligent Manslaughter (total deaths)', ylabel='Life Expectancy')

# Gun Murder 
s2=sns.regplot(x='GunMurderTot', y='LifeExp2018', data=df, ax=ax2)
s2.set(title='Corr: {:.4f}'.format(corr_GunMurder), xlabel='Gun Murder (total deaths)', ylabel='')

# Murders
s3=sns.regplot(x='MurderTot', y='LifeExp2018', data=df, ax=ax3)
s3.set(title='Corr: {:.4f}'.format(corr_MurderTot), xlabel='Murders (total deaths)', ylabel='')

fig.suptitle('Life Expectancy per state (2018)')
plt.savefig('regplots_crime.png')
plt.show()
```

```python
# Regression plot for gun ownership
```

```python
# Regression plots for murder (rates)
```

```python
corr_MurderNMRate=corr_matrix_murder.loc['LifeExp2018', 'MurderNMRate']
corr_GunMurderRate=corr_matrix_murder.loc['LifeExp2018', 'GunMurderRate']
corr_MurderRate=corr_matrix_murder.loc['LifeExp2018', 'MurderRate']
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(nrows=1,ncols=3, sharey=True, figsize=(17,4))

# Non-Negligent mansalughter Murder rate
s1=sns.regplot(x='MurderNMRate', y='LifeExp2018', data=df, ax=ax1)
s1.set(title='Corr: {:.4f}'.format(corr_MurderNMRate), xlabel='Murder and non-negligent manslaughter (rate)', ylabel='Life Expectancy')

# Gun Murder rate
s2=sns.regplot(x='GunMurderRate', y='LifeExp2018', data=df, ax=ax2)
s2.set(title='Corr: {:.4f}'.format(corr_GunMurderRate), xlabel='Gun Murder (rate)', ylabel='')

# Murder total

s3=sns.regplot(x='MurderRate', y='LifeExp2018', data=df, ax=ax3)
s3.set(title='Corr: {:.4f}'.format(corr_MurderRate), xlabel='Murders (rate)', ylabel='')

fig.suptitle('Life Expectancy per state (2018)')
plt.savefig('regplots_Murder_rates_crime.png')
plt.show ()
```

```python
# ECDF for murder (totals)

x1=np.sort(df['MurderNMTot'])
x2=np.sort(df['GunMurderTot'])
x3=np.sort(df['MurderTot'])
y=np.arange(1, len(x1)+1)/len(x1)
fig, ax=plt.subplots ()
ax.plot(x1,y,'.',linestyle='none',color='blue',label='Murder NM')
ax.plot(x2,y,'.',linestyle='none',color='green',label='Gun Murder')
ax.plot(x3,y,'.',linestyle='none',color='red',label='Murder')

ax.set_xlabel('Murder (tot)')
ax.set_ylabel('ECDF')
ax.set_title('Empirical Cummulative Distribution \n of total murders')
ax.legend()
plt.margins(0.02)
plt.savefig('ecdf_murder.png')
plt.show ()

```

```python
 # ECDF for murder rates
x1=np.sort(df['MurderNMRate'])
x2=np.sort(df['GunMurderRate'])
x3=np.sort(df['MurderRate'])
y=np.arange(1, len(x1)+1)/len(x1)
fig, ax=plt.subplots ()
ax.plot(x1,y,'.',linestyle='none',color='blue',label='Murder NM')
ax.plot(x2,y,'.',linestyle='none',color='green',label='Gun Murder')
ax.plot(x3,y,'.',linestyle='none',color='red',label='Murder')

ax.set_xlabel('Murder (tot)')
ax.set_ylabel('ECDF')
ax.set_title('Empirical Cummulative Distribution \n of murders rates')
ax.legend()
plt.margins(0.02)
plt.savefig('ecdf_murder_rates.png')
plt.show ()
```
